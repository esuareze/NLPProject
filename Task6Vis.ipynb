```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open("QuickData/KwToTitleAugmented.data", "rb") as f:
    kwToTitle = pickle.load(f)
```

```python
concepts = kwToTitle.keys()
avgTitleInKw = [value['avgPTitleInKw'] for key, value in kwToTitle.items()]
totalTitleInKw = [value['totalPTitleInKw'] for key, value in kwToTitle.items()]
avgKwInTitle = [value['avgPKwInTitle'] for key, value in kwToTitle.items()]
totalKwInTitle = [value['totalPKwInTitle'] for key, value in kwToTitle.items()]
pArticlesWithKw =[value['pArticlesWithKw'] for key, value in kwToTitle.items()]
X = concepts
  
X_axis = np.arange(len(X))
plt.figure("figSize", (20,10))
plt.bar(X_axis - 0.2, avgTitleInKw, 0.2, label = 'avgTitleInKw')
plt.bar(X_axis - 0, totalTitleInKw, 0.2, label = 'totalTitleInKw')
plt.bar(X_axis + 0.2, avgKwInTitle, 0.2, label = 'avgKwInTitle')
plt.bar(X_axis + 0.4, totalKwInTitle, 0.2, label = 'totalKwInTitle')
  
plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Number of Students")
plt.title("Number of Students in each group")
plt.legend()
plt.show()

```
