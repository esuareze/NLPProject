```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open("QuickData/topConceptCounts", "rb") as f:
    topKwCounts = pickle.load(f)
```

```python
months = ["Dec-2019", "Jan-2020", "Feb-2020", "Mar-2020", "Apr-2020", "May-2020", "Jun-2020", "Jul-2020", "Ag-2020", "Sep-2020", "Oct-2020", "Nov-2020", "Dec-2020", "Jan-2021", "Feb-2021", "Mar-2021", "Apr-2021", "May-2021", "Jun-2021", "Jul-2021", "Ag-2021", "Sep-2021"]
print(topKwCounts.keys())
plt.figure("figSize", (20,10))
for k in topKwCounts.keys():
    plt.plot(months, topKwCounts[k], label=k)
plt.legend()
plt.show()
```

```python
import pickle
with open("articlesData.data","rb") as f:
    articles = pickle.load(f)
numMonths = articles[-1]['monthInd'] +1
ccsArticlesCounts = [0]*numMonths
for a in articles:
    ccsArticlesCounts[a['monthInd']] += 1 if len(a['ccsClass']) > 0 else 0
```

```python
f, ax = plt.subplots(1, figsize=(20,10))

ax.plot(months, ccsArticlesCounts)
ax.set_title('amount of articles that are classified')
ax.set_ylim(ymin=0)
plt.show()
```
