```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open("QuickData/vec2wordoverlapping1", "rb") as f:
    kwToAbstract = pickle.load(f)
with open("QuickData/vec2wordoverlapping1raw", "rb") as f:
    kwToAbstractRaw = pickle.load(f)
```

```python
sorted_by_quantity = sorted(kwToAbstractRaw.items(), key=lambda x: len(x[1]), reverse=True)
top100names = [x[0] for x in sorted_by_quantity]



sorted_results = sorted(kwToAbstract.items(), key=lambda x: x[1], reverse=True)
names = [x[0] for x in sorted_results if x[0] in top100names]
values = [x[1] for x in sorted_results if x[0] in top100names]
plt.figure("figSize", (20,10))
plt.bar(x=names, height=values)
plt.xticks(rotation=90)
plt.show()
```
