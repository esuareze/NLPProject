```python
import pickle
with open("articlesData.data","rb") as f:
    articles = pickle.load(f)
```

```python
numMonths = articles[-1]['monthInd'] +1
kwCounts = [0]*numMonths
kwArticlesCounts = [0]*numMonths
for a in articles:
    kwCounts[a['monthInd']]+= len(a['authorTags'])
    kwArticlesCounts[a['monthInd']] += 1 if len(a['authorTags']) > 0 else 0
```

```python
import matplotlib.pyplot as plt
f, ax = plt.subplots(1)
ax.plot(kwCounts)
ax.set_ylim(ymin=0)
plt.show()
```

```python
plt.plot(kwArticlesCounts)
plt.show()
```
