```python
with open("QuickData/num_results", "r") as f:
    num_results = f.readline()[:-1] #remove the line jump
num_results = num_results.split(",")
```

```python
num_results = [int(x) for x in num_results]
```

```python
import matplotlib.pyplot as plt
months = ["Dec-2019", "Jan-2020", "Feb-2020", "Mar-2020", "Apr-2020", "May-2020", "Jun-2020", "Jul-2020", "Ag-2020", "Sep-2020", "Oct-2020", "Nov-2020", "Dec-2020", "Jan-2021", "Feb-2021", "Mar-2021", "Apr-2021", "May-2021", "Jun-2021", "Jul-2021", "Ag-2021", "Sep-2021"]

f, ax = plt.subplots(1, figsize=(20,10))

ax.plot(months, num_results)
ax.set_title('search results')
ax.set_ylim(ymin=0)
plt.show()
```
